```python
# Recolter le texte de la critique
import urllib
from urllib.request import urlopen
from bs4 import BeautifulSoup as beautifulsoup
import requests as reqt

url="https://www.jeuxvideo.com/"

reponse = reqt.get(url)

soupe = beautifulsoup(reponse.text, "lxml")

print (soupe.title)

print (reponse.text)
```

```python
# Voir la note du Jeux video
url="https://www.jeuxvideo.com/jeux/ps4/jeu-412345/"
pageweb = urlopen(url)
soupe = beautifulsoup(pageweb.read(), "lxml")
notation = (soupe.find("div", {"class": "gameCharacteristicsMain__gauge"})).text
print(notation)
#print(soupe.html.div)
```

```python
# Voir sa description
synopsis = soupe.find("p", {"class": "gameCharacteristicsMain__synopsis"}).text
print(synopsis)
```

```python
# Voir l'intro de la critique
url2="https://www.jeuxvideo.com/test/1237466/the-last-of-us-part-ii-la-vengeance-au-coeur-d-un-recit-epoustouflant.htm"
page_critique = urlopen(url2)
soupe2 = beautifulsoup(page_critique.read(), "lxml")
critique = soupe2.find("p",{"class": "intro-article"}).text
print(critique)
```

```python
# l'integralité de la critique
contenu_critique = soupe2.find_all("p")
print(contenu_critique)
```
